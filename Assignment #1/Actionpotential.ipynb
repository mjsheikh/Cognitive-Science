```python
import  matplotlib.pyplot as plt
def fn(x,y,alp):
    if(x<=0):
        return alp/(1-x)+y
    elif(0<x<alp+y):
        return alp+y
    elif(x>=alp+y):
        return -1
```

```python
def cal(xold,yold,sig,mu,alp):
    xnew=fn(xold,yold,alp)
    ynew=yold-mu*(xold+1)+mu*sig
    return xnew,ynew
```

## 1) The shape of the nonlinear function f (x,y), plotted for f(x,y) for alp=6.0 and y=3.93

```python
import numpy as np
l=np.linspace(-2,2,20)
y=-3.93
alp=6.0
plt.plot(l,[fn(i,y,alp)for i in l] )
plt.xlabel(r'$x_n$ ', fontsize=20)
plt.ylabel(r'$x_{n+1}$ ',fontsize=20)
plt.show()
```

```python
def neuronplot(x=-1,y=-1,sig=-0.1,mu=0.001,alp=4):
    t=list()
    for i in range(2000):
        xnew,ynew=cal(x,y,sig,mu,alp)
        x=xnew
        y=ynew
        t.append(x)
    plt.figure(figsize=(10,5))
    plt.plot([i for i in range(2000)],t)
    plt.show()
```

# 2) Wave forms of neuronal action potential

### a)  Transition to the regime of silence with alp=4  and sig=-0.1

```python
neuronplot(x=-1,y=-2.92,sig=-0.1,mu=0.001,alp=4)
plt.savefig('fn.png', bbox_inches='tight')
```

### b) The regimes of continuous tonic spiking  mapped for alp=4 and sig=0.01

```python
#-----------------------------------continuous tonic spiking--------------------------------
neuronplot(x=-1,y=-2.99,sig=0.01,mu=0.001,alp=4)
```
